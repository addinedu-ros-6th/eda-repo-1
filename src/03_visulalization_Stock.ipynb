#  고영,큐렉소,미래 컴퍼니 주가 평균

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
# List of companies in the robot + autonomous driving sector
# Example: iRobot (IRBT), Tesla (TSLA), NVIDIA (NVDA), etc.
companies = ['098460.kq', '060280.kq', '049950.kq']
# Fetch historical data from Yahoo Finance
start_date = "2020-01-01"
end_date = "2024-05-30"
data = yf.download(companies, start=start_date, end=end_date)['Adj Close']
# Handling missing data by forward filling
data.fillna(method='ffill', inplace=True)
# Normalize the data to start from the same point for index calculation
normalized_data = data / data.iloc[0]
# Calculate the equal-weighted sector index
sector_index = normalized_data.mean(axis=1)
# Plot the sector index
plt.figure(figsize=(14, 7))
plt.plot(sector_index.index, sector_index, label='SUGICAL ROBOT COMPANIES Index')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('SUGICAL ROBOT COMPANIES Index (2020-2024)')
plt.legend()
plt.grid(True)
plt.show()
```

## 한국 기업별 주가

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib 

# 로봇 및 자율주행 섹터에 속한 회사들의 목록 (대문자로 수정)
companies = ['098460.KQ', '060280.KQ', '049950.KQ'] 

# 회사 이름을 한글로 매칭
company_names_kr = {
    '098460.KQ': '고영',
    '060280.KQ': '큐렉소',
    '049950.KQ': '미래컴퍼니'
}

# Yahoo Finance에서 역사적 데이터를 다운로드합니다.
start_date = "2020-01-01"
end_date = "2024-05-30"
data = yf.download(companies, start=start_date, end=end_date)['Adj Close']

# 다운로드된 데이터의 열 이름을 출력하여 확인
print("Downloaded data columns:", data.columns)

# 결측 데이터 처리 (앞의 값을 채워 넣음)
data.ffill(inplace=True)

# 개별 주식의 정규화 데이터를 계산
normalized_data = data / data.iloc[0]

# 정규화된 데이터의 열 이름을 출력하여 확인
print("Normalized data columns:", normalized_data.columns)

# 개별 주식을 그리기 위한 그래프 설정
plt.figure(figsize=(14, 7))

# 각 회사의 정규화된 주식 가격을 플롯
for company in companies:
    if company in normalized_data.columns:
        plt.plot(normalized_data.index, normalized_data[company], label=company_names_kr[company])
    else:
        print(f"Warning: {company} not found in normalized data columns")

# 그래프 제목 및 축 레이블 설정
plt.xlabel('날짜')
plt.ylabel('정규화된 가격')
plt.title('한국 수술 로봇 (2020-2024)')
plt.legend()
plt.grid(True)
plt.show()
```

# 한국기업과 미국기업 평균주가 비교

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# List of companies in the surgical robots sector
US = ['ISRG', 'SYK', 'MDT', 'GMED','ASXC','SNN', 'JNJ']
KR = ['098460.kq', '060280.kq', '049950.kq']

# Fetch historical data from Yahoo Finance
start_date = "2020-01-01"
end_date = "2024-06-30"
us_data = yf.download(US, start=start_date, end=end_date)['Adj Close']
kr_data = yf.download(KR, start=start_date, end=end_date)['Adj Close']

# Handling missing data by forward filling
us_data.fillna(method='ffill', inplace=True)
kr_data.fillna(method='ffill', inplace=True)

# Normalize the data to start from the same point for index calculation
us_normalized_data = us_data / us_data.iloc[0]
kr_normalized_data = kr_data / kr_data.iloc[0]

# Calculate the equal-weighted sector index
us_sector_index = us_normalized_data.mean(axis=1)
kr_sector_index = kr_normalized_data.mean(axis=1)


# Plot the sector index
plt.figure(figsize=(14, 7))
plt.plot(us_sector_index.index, us_sector_index, color = 'red', label='US')
plt.plot(kr_sector_index.index, kr_sector_index, color = 'blue', label='KR')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Surgical Robot Sector Index (2020.01-2024.05)')
plt.legend()
plt.grid(True)
plt.show()
```
