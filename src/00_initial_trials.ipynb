## 미국 의료용 로봇 섹터 근 5년간 평균 종가 비교

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
# List of companies in the surgical robots sector
US = ['ISRG', 'SYK', 'MDT', 'GMED', 'ASXC', 'SNN', 'JNJ']
# Fetch historical data from Yahoo Finance
start_date = "2019-06-01"
end_date = "2024-05-30"
data = yf.download(US, start=start_date, end=end_date)['Adj Close']
# Handling missing data by forward filling
data.fillna(method='ffill', inplace=True)
# Normalize the data to start from the same point for index calculation
normalized_data = data / data.iloc[0]
# Calculate the equal-weighted sector index
us_sector_index = normalized_data.mean(axis=1)

# Plot the sector index
plt.figure(figsize=(14, 7))
plt.plot(us_sector_index.index, us_sector_index, color = 'blue', label='Surgical Robot Sector Index')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Robot + Autonomous Driving Sector Index (2014-2024)')
plt.legend()
plt.grid(True)
plt.show()
```

## 한국 vs 미국 의료용 로봇 섹터 평균 종가 비교

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# List of companies in the surgical robots sector
US = ['ISRG', 'SYK', 'MDT', 'GMED', 'ASXC', 'SNN', 'JNJ']
KR = ['098460.kq', '060280.kq', '049950.kq']

# Fetch historical data from Yahoo Finance
start_date = "2020-01-01"
end_date = "2023-03-30"
us_data = yf.download(US, start=start_date, end=end_date)['Adj Close']
kr_data = yf.download(KR, start=start_date, end=end_date)['Adj Close']

# Handling missing data by forward filling
us_data.fillna(method='ffill', inplace=True)
kr_data.fillna(method='ffill', inplace=True)

# Normalize the data to start from the same point for index calculation
us_normalized_data = us_data / us_data.iloc[0]
kr_normalized_data = kr_data / kr_data.iloc[0]

# Calculate the equal-weighted sector index
us_sector_index = us_normalized_data.mean(axis=1)
kr_sector_index = kr_normalized_data.mean(axis=1)


# Plot the sector index
plt.figure(figsize=(14, 7))
plt.plot(us_sector_index.index, us_sector_index, color = 'red', label='US')
plt.plot(kr_sector_index.index, kr_sector_index, color = 'blue', label='KR')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Surgical Robot Sector Index (2023.02.01-2023.06.30)')
plt.legend()
plt.grid(True)
plt.show()
```

# pytrends를 이용한 트렌드 검색 시각화?

구글에 검색됐던 키워드의 검색 횟수(빈돗수)를 기준으로 자료 제공 중. 인기도 비교할 때 유용. 그러나 우리 케이스엔 맞지 않음

```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
keyword1 = "고영"
keyword2 = "큐렉소"
keyword3 = "미래컴퍼니"
period = "today 5-y"
trend_obj = TrendReq()
trend_obj.build_payload(kw_list=[keyword1, keyword2, keyword3], timeframe=period, geo="KR")
trend_df = trend_obj.interest_over_time()
plt.rc('font', family='NanumGothic')
plt.style.use("ggplot")
plt.figure(figsize=(14,5))
trend_df[keyword1].plot()
trend_df[keyword2].plot()
trend_df[keyword3].plot()
plt.title("고영, 큐렉소, 미래컴퍼니의 검색 트렌드 분석결과", size=15)
plt.legend(loc="best")
plt.show()
```
